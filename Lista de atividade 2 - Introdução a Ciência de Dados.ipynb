18. Elabore uma função chamada calcula_desconto que receba dois parâmetros: o valor de um produto e a porcentagem de desconto. A função deve retornar o valor final do produto com o desconto aplicado. Teste a função e imprima o resultado.


```python
def calcula_desconto(valor, porcentagem):
  
    desconto = valor * (porcentagem / 100)
    valor_final = valor - desconto
    return valor_final

preco_original = 150.00
percentual_off = 15

resultado = calcula_desconto(preco_original, percentual_off)

print(f"Valor original: R$ {preco_original:.2f}")
print(f"Desconto: {percentual_off}%")
print(f"Valor final com desconto: R$ {resultado:.2f}")
```

20. Elabore um código que receba uma lista de números inteiros mistos (positivos, negativos e zeros). Utilize um laço for com estruturas condicionais if/elif/else para contabilizar quantos números são positivos, quantos são negativos e quantos são exatamente zero. Ao final, imprima esse resumo.


```python
numeros = [10, -5, 0, 7, -22, 0, 15, -3, 0, 42]

positivos = 0
negativos = 0
zeros = 0

for num in numeros:
    if num > 0:
        positivos += 1
    elif num < 0:
        negativos += 1
    else:
        zeros += 1

print("==========================")
print("--- Resumo da Contagem ---")
print(f"Positivos: {positivos}")
print(f"Negativos: {negativos}")
print(f"Zeros:     {zeros}")
```
