```python
nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))
nota3 = float(input("Digite a terceira nota: "))
nota4 = float(input("Digite a quarta nota: "))

media = (nota1 + nota2 + nota3 + nota4) / 4

print("A média do Bimestre é:", media)
```

# Atividade 9 - Transforme e mostre a temperatura em graus Celsius.

O Código a seguir transforma a temperatura Fahrenheit em Celsius.

```python
temperatura_fahrenheit = float(input("Digite a temperatura em Fahrenheit: "))
temperatura_celsius = (temperatura_fahrenheit - 32) * 5.0/9.0

print(temperatura_fahrenheit, "graus Fahrenheit equivalem a", temperatura_celsius, "graus Celsius.")
```

# Atividade 10 - Temperatura em graus Celsius, transforme e mostre em graus Fahrenheit.

O Código a seguir transforma uma temperatura Celsius em graus Fahrenheit.

```python
temperatura_celsius = float(input("Digite a temperatura em Celsius: "))
temperatura_fahrenheit = (temperatura_celsius * 9.0/5.0) + 32

print(temperatura_celsius, "graus Celsius equivalem a", temperatura_fahrenheit, "graus Fahrenheit.")
```

# Atividade 11 - Números inteiros e um número real, calcule e mostre.

# Atividade 12 - altura de uma pessoa, e informe qual seria o peso ideal.

O Código a seguir calcula o peso ideal de uma pessoa de acordo com sua altura.

```python
altura = float(input("Digie sua altura em metros: "))
sexo = input("Digite seu sexo (M para masculino, F para feminino): ").upper()
peso_ideal_homem = (72.7 * altura) - 58
peso_ideal_mulher = (62.1 * altura) - 44.7

if sexo == "M":
    print("O peso ideal para a sua altura é:", peso_ideal_homem, "kg")
elif sexo == "F":
    print("O peso ideal para a sua altura é:", peso_ideal_mulher, "kg")
else:
    print("Sexo inválido. Por favor, digite M para masculino ou F para feminino.")
```
