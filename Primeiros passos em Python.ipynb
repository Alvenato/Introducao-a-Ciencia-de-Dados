"SegundosDia" é uma variavel que recebe um calculo matemático e depois imprime na tela!

```python
dia = 24
hora = 60
minuto = 60 

segundosDia = dia*hora*minuto
print("Seu dia tem o total de: ", segundosDia, "segundos!") 
```

"dia", "hora" e "minuto" são variaveis que estão atribuidas a algum valor declarado, "segundosDia" recebe o calculo entre as variaveis declaradas, depois imprime em tela a mensagem "Seu dia tem o total de: "variavel", segundos!.
