```python
nome = input("Digite seu nome: ")
anonasc = int(input("Digite o ano do seu nascimento: "))
hoje = int(input("Digite o ano atual: "))

idadeatual = hoje - anonasc
print("=====================================")
print("Seu nome é: ",nome)
print("A idade atual é: ",idadeatual,"anos")
print("=====================================")
print("=====================================")
print("O usuario {} tem {} anos de idade.".format(nome,idadeatual))
print("=====================================")
```
